```python
# Temporal Analysis by using seaborn method
temporal_query = """
Select Date_Occ
From crime_analysis.crime_data
"""
temporal_data = pd.read_sql(temporal_query, conn)
temporal_data.head()
temporal_data['Date_Occ'] = pd.to_datetime(temporal_data['Date_Occ'])   

# Aggregate the data by date and also count the number of crimes on each date

temporal_data['Count'] = 1  # Adding a count column for aggregation
temporal_data = temporal_data.resample('ME', on='Date_Occ').sum().reset_index() 

import seaborn as sns
plt.figure(figsize=(10, 5))
sns.lineplot(x='Date_Occ', y='Count', data=temporal_data)
plt.title('Trends in Crime Occurrence Over Time')
plt.xlabel('Date_Occ')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
# Spacial Analysis
query = """
Select Lat,Lon
From crime_analysis.crime_data
"""

df = pd.read_sql(query, conn)


import folium

crime_map = folium.Map(location=[df['Lat'].mean(),df['Lon'].mean()], zoom_start=10)

for index, row in df.iterrows():
    folium.Marker([row['Lat'], row['Lon']]).add_to(crime_map)

crime_map
```

```python
# Victim  Analysis
# 1.Victim Analysis based on Age:

vict_age_query = """
Select Vict_Age
From crime_data
"""

vict_age_df = pd.read_sql(vict_age_query, conn)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(vict_age_df['Vict_Age'], bins=12, color='green', edgecolor='black')
plt.title('Distribution of Victim Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
# 2.Victim Analysis based on gender  by using Histogram Graph
Vict_gender_query = " Select Vict_Sex from crime_analysis.crime_data "


victim_data_df = pd.read_sql(Vict_gender_query, conn)

plt.figure(figsize=(8, 6))
plt.hist(victim_data_df['Vict_Sex'], color='salmon', edgecolor='black')
plt.title('Distribution of Victim Genders')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
#  Crime rates based on gender using Scatter Plot method:

crime_gender_query = """
Select Vict_Sex, COUNT(*) AS Crime_Count
From crime_data
Group By Vict_Sex
"""

crime_gender_data = pd.read_sql(crime_gender_query, conn)

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Vict_Sex', y='Crime_Count', data=crime_gender_data,marker = 'H', s=800,edgecolor='black',facecolor='red')
plt.title('Difference in Crime Rates Between Genders')
plt.xlabel('Gender')
plt.ylabel('Number of Crimes')
plt.grid(False)
plt.show()
```

```python
# Crime Rates between Male and Female Victim using Plotly Pie Method
gender_query = """
Select Vict_Sex
From crime_data
Where Vict_Sex IN ('F', 'M')
"""

gender_data = pd.read_sql(gender_query, conn)

crime_counts = gender_data['Vict_Sex'].value_counts()

plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
plt.pie(crime_counts, labels=crime_counts.index, autopct='%1.1f%%', colors=['blue', 'orange'])
plt.title('Difference in Crime Rates Between Genders')
plt.axis('equal') 
plt.show()
```

```python
# Investagiation Status Analysis by using BarPLot Method 

status_query = """
Select Status
From crime_analysis.crime_data
"""
status_data = pd.read_sql(status_query, conn)

status_counts = status_data['Status'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=status_counts.index, y=status_counts.values, palette="viridis")
plt.title('Status of Reported Crimes')
plt.xlabel('Status')
plt.ylabel('No.of Crimes')
plt.xticks(rotation=45)# Rotate x-axis labels for better readability
plt.tight_layout()
plt.grid(False)
plt.show()
```
